```python
#New data with most affecting factors

df=df[['Country', 'Year','Status', 'Adult Mortality','infant deaths','Alcohol', 'percentage expenditure','Measles ',' BMI ',
'under-five deaths ', 'Polio','Diphtheria ', ' HIV/AIDS', 'GDP',' thinness  1-19 years', ' thinness 5-9 years',
'Income composition of resources', 'Schooling','Life expectancy ']]

df.head()
```

```python
#df['Status'].unique()
status={"Developing":1,"Developed":2}
df['Status']=df['Status'].replace(status)
```

```python
df.isna().sum()
```

```python
#Fill missing data

df['Adult Mortality']=df['Adult Mortality'].fillna(df['Adult Mortality'].mean())
df['Alcohol']=df['Alcohol'].fillna(df['Alcohol'].mean())
df[' BMI ']=df[' BMI '].fillna(df[' BMI '].mean())
df['Polio']=df['Polio'].fillna(df['Polio'].mean())
df['Diphtheria ']=df['Diphtheria '].fillna(df['Diphtheria '].mean())
df[' thinness  1-19 years']=df[' thinness  1-19 years'].fillna(df[' thinness  1-19 years'].mean())
df[' thinness 5-9 years']=df[' thinness 5-9 years'].fillna(df[' thinness 5-9 years'].mean())
df['Schooling']=df['Schooling'].fillna(df['Schooling'].mean())
df['Income composition of resources']=df['Income composition of resources'].fillna(df['Income composition of resources'].mean())
df['Life expectancy ']=df['Life expectancy '].fillna(df['Life expectancy '].mean())
df=df.dropna()
df.isna().sum()
```

**Handeling Imbalanced data**

```python
sns.countplot(df['Status'])
```

```python
#under sampling of Dataset
#1->developing--------more data available
#2-> developed--------less data available
developing=df[df["Status"]==1].index
developed=df[df["Status"]==2].index   
random_index=np.array(np.random.choice(developing,len(developed),replace=False))
udersample_index=np.concatenate([developed,random_index])
undersample=df.loc[udersample_index,:]

sns.countplot(undersample['Status'])
```

```python
#over sampling of Dataset
#1->developing--------more data available
#2-> developed--------less data available
developing=df[df["Status"]==1].index
developed=df[df["Status"]==2].index   
random_index=np.array(np.random.choice(developed,len(developing),replace=True))
oversample_index=np.concatenate([developing,random_index])
oversample=df.loc[udersample_index,:]

sns.countplot(oversample['Status'])
```

**Removing Outliers**

IQR method

```python
df=oversample
df.head(5)
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df)
```

```python
q3=df.quantile(0.75)
q1=df.quantile(0.25)
IQR=q3-q1

#limits
upper_limit=q1-1.5*IQR
lower_limit=q3+1.5*IQR
```

```python
df = df[~((df < (q1 - 1.5 * IQR)) |(df > (q3 + 1.5 * IQR))).any(axis=1)]
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df)
```
