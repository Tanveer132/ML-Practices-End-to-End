**Data visualization :**

    It gives clear idea of data and becomes easier to identify trends, patterns and outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/MelaSales.csv')
df.head()
```

```python
plt.plot(df['Week1'],marker="*",linestyle="--",color="r")
```

```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot(df['Week1'],color='r')
ax[0,1].plot(df['Week2'],color='g')
ax[1,0].plot(df['Week3'],color='b')
ax[1,1].plot(df[['Week1','Week2','Week3']])
```

```python
df.plot(kind='bar',x='Day',linestyle='--',edgecolor='k')
plt.ylabel('')
plt.title('Weekly sells')
```

```python
df=pd.read_csv('/content/medals_by_country_2016.csv')
df.head()
```

```python
fig,ax=plt.subplots()
ax.bar(df['Country'],df['Gold'],label="Gold",color='r')
ax.bar(df['Country'],df['Silver'],bottom=df['Gold'],label="Silver",color='g')
ax.bar(df['Country'],df['Bronze'],bottom=df['Gold']+df['Silver'],label="Bronze",color='b')
ax.set_xticklabels(df['Country'],rotation=90)
ax.set_title('Medals by country')
```

```python
df['Gold'].plot(kind='hist')
```

```python
df.plot(kind='hist',bins=[10,20,30,40,50,60],fill=False,hatch='o')
```

```python
df.plot(kind='hist',bins=[10,20,30,40,50,60],histtype='step')
```

```python
df=pd.read_csv('/content/world-happiness-report.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
#boxplot to identify outliers
plt.figure(figsize=(10,6))
df.iloc[:,2:].boxplot()
```

```python
#scatterplot
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.scatterplot(x=df['Freedom to make life choices'],y=df['Positive affect'])
plt.subplot(2,2,2)
sns.scatterplot(x=df['Perceptions of corruption'],y=df['Negative affect'])
plt.subplot(2,2,3)
sns.scatterplot(x=df['Social support'],y=df['Positive affect'])
plt.subplot(2,2,4)
sns.scatterplot(x=df['Generosity'],y=df['Positive affect'])
```

```python

india=df[df['Country name']=='India']
india.head()
```

```python
india.info()
```

```python
sns.lineplot(x=india['year'],y=india['Log GDP per capita'])
```

```python
plt.figure(figsize=(15,7))
plt.subplot(2,2,1)
sns.barplot(x=india['year'],y=india['Healthy life expectancy at birth'])
plt.subplot(2,2,2)
sns.barplot(x=india['year'],y=india['Social support'])
plt.subplot(2,2,3)
sns.barplot(x=india['year'],y=india['Perceptions of corruption'])
plt.subplot(2,2,4)
sns.barplot(x=india['year'],y=india['Life Ladder'])
```

```python
sns.pairplot(india)
```

```python
sns.heatmap(india.corr())
```
