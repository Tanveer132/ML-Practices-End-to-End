#Principal Component Analysis

 A more common way of speeding up a machine learning algorithm is by using Principal Component Analysis (PCA). If your learning algorithm is too slow because the input dimension is too high, then using PCA to speed it up can be a reasonable choice. This is probably the most common application of PCA. Another common application of PCA is for data visualization.
 
Principal component analysis (PCA) is a technique for reducing the dimensionality of such datasets, increasing interpretability but at the same time minimizing information loss.

```python
import numpy as np
from numpy.linalg import eig
from sklearn.decomposition import PCA
```

##Using numpy

```python
data=np.array([[1,2],[3,4],[5,6]])
data
```

```python
# get scaled data
m=np.mean(data,axis=0)

scaled_data=m-data
scaled_data
```

```python
# get covarience matrix
V=np.cov(scaled_data.T)
V 
```

```python
# get eigen vectors and eigen values
values,vectors=eig(V)
```

```python
values.T
```

```python
vectors.T
```

```python
# get transformed data
P = vectors.T.dot(scaled_data.T)
P.T
```

**Conclusion :** Here dimensions reduced from 2 to 1

##Using pca in sklearn

```python
pca=PCA()
pca.fit(data)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
P=pca.transform(data)
P
```
